```python
#I need to prepare the data.  First I will recalculate distance and speed again
from math import radians, cos, sin, asin, sqrt

# step: you need to calculate the distance
def calculate(startstationlongitude,startstationlatitude,endstationlongitude,endstationlatitude):
    
    startstationlongitude=map(radians,np.array(startstationlongitude))
    startstationlatitude=map(radians,np.array(startstationlatitude))
    
    endstationlongitude=map(radians,np.array(endstationlongitude))
    endstationlatitude=map(radians,np.array(endstationlatitude))
    
    startstationlongitude=np.array(list(startstationlongitude)).reshape(-1,1)
    endstationlongitude=np.array(list(endstationlongitude)).reshape(-1,1)
    startstationlatitude=np.array(list(startstationlatitude)).reshape(-1,1)
    endstationlatitude=np.array(list(endstationlatitude)).reshape(-1,1)
    
    dlon=endstationlongitude-startstationlatitude
    dlat=endstationlatitude-startstationlatitude
    # haversine formula
    a=np.sin(dlat/2)**2+np.cos(startstationlatitude)*np.cos(endstationlatitude)*np.sin(dlon/2)**2
    c=2*np.arcsin(np.sqrt(a))
    r=6371
    distance=c*r*1000
    return distance

df["distance"]=calculate(df["startstationlatitude"],df["startstationlatitude"],df["endstationlongitude"],df["endstationlatitude"])
```
