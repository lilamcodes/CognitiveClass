```python
#statement to display load status

def read_and_prepare_data(load_prepared_data=False, save_data=True):

    if (load_prepared_data):
        print("Loading 2017-prepared_data.csv ...")
        df = pd.read_csv("2017-prepared_data.csv")
        print("successfully.")
        return df
    
# Go through monthly zipped csv files and imported into dataframe. 
# Append all those files into one dataframe
# Added in column edits after reviewing initial outputs
    n_months = 12
    for month in range(1,n_months+1):
        filename = "2017"+str(month).zfill(2)+"-citibike-tripdata.csv.zip"
        print("Loading: "+filename)
        dm = pd.read_csv(filename)
        dm.columns = [x.lower() for x in dm.columns]
        dm.columns = dm.columns.str.replace(' ', '') 
        if (month == 1):
            df = dm
        else:
            df = pd.concat([df, dm])
            
#all csv files combined and saved to df
    if (save_data):
        fnameout = "2017-prepared_data.csv"
        df.to_csv(fnameout, index=False)
        print('\nData saved to ' + fnameout)
        
    return df

```
